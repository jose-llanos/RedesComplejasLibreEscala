```python
import os,glob,pandas as pd,numpy as np,matplotlib.pyplot as plt
```

```python
#listfolder=['Funcion_Ackley','Funcion_Beale','Funcion_Camel','Funcion_Esfera']
listfolder=['Funcion_Esfera']
for i in listfolder:
    listtxt=glob.glob(i+'/ResultadosTXT/*.txt')
#listtxt = ['Funcion_Esfera/ResultadosTXT/Esfera_barabasi_best.txt', 
#           'Funcion_Esfera/ResultadosTXT/Esfera_barabasi.txt',
#           'Funcion_Esfera/ResultadosTXT/Esfera_evolutive.txt',
#           'Funcion_Esfera/ResultadosTXT/Esfera_sw_best.txt',
#           'Funcion_Esfera/ResultadosTXT/Esfera_sw.txt',
#          ]
df=pd.DataFrame()
plt.figure(dpi=300)
for j in listtxt:
    archivo=open(j,"r")
    lineas=archivo.readlines()
    line=lineas[0].split(',')
    df[os.path.basename(j)[:-4]] = line
    df[os.path.basename(j)[:-4]] = pd.to_numeric(df[os.path.basename(j)[:-4]], errors='coerce')
    plt.loglog(range(1,len(df)+1),df[os.path.basename(j)[:-4]],label=os.path.basename(j)[:-4])
plt.xlabel('Iteration')
plt.ylabel('Fitness')
plt.title(i)
#plt.title('Function Sphere')
plt.legend()
plt.grid(ls=':')
plt.savefig(i+".jpg",dpi=300)
#plt.savefig("Function_Esfera_100.jpg",dpi=300)
plt.show()
```
