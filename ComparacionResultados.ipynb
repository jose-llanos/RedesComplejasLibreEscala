```python
import os,glob,pandas as pd,numpy as np
```

```python
listfolder=['Funcion_Beale','Funcion_Ackley']
for i in listfolder:
    listtxt=glob.glob(i+'/ResultadosTXT/*.txt')
    df=pd.DataFrame()
    for j in listtxt:
        archivo=open(j,"r")
        lineas=archivo.readlines()
        line=lineas[0].split(',')
        df[os.path.basename(j)[:-4]] = line
        df[os.path.basename(j)[:-4]] = pd.to_numeric(df[os.path.basename(j)[:-4]], errors='coerce')
    df.head
    df.plot()

```
