```python
import os,glob,pandas as pd,numpy as np,matplotlib.pyplot as plt
```

```python
listfolder=['Graficas']
for i in listfolder:
    listtxt=glob.glob(i+'/ResultadosTXT/*.txt')
    df=pd.DataFrame()
    plt.figure(dpi=100)
    for j in listtxt:
        archivo=open(j,"r")
        lineas=archivo.readlines()
        line=lineas[0].split(',')
        df[os.path.basename(j)[:-4]] = line
        df[os.path.basename(j)[:-4]] = pd.to_numeric(df[os.path.basename(j)[:-4]], errors='coerce')
        plt.loglog(range(1,len(df)+1),df[os.path.basename(j)[:-4]],label=os.path.basename(j)[:-4])
    plt.xlabel('Iteración')
    plt.ylabel('Distancia')
    plt.title(i)
    plt.legend()
    plt.grid(ls=':')
    plt.savefig(i+".jpg",dpi=300)
    plt.show()
```
