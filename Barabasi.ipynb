**Importo la libreria NetworkX de Python, que sirve para el estudio de gráficos y análisis de redes**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

**Genero la gráfica para la red Barabasi**

```python
# Número de nodos
n = 100
# Número de aristas
m = 3
# Semilla
s = 0

# Genero la gráfica de barabasi
G = nx.barabasi_albert_graph(n, m, s)
nx.draw(G, node_size=1)
```

```python
degrees = dict(G.degree())
degree_values = sorted(set(degrees.values()))
histogram = [list(degrees.values()).count(i) / float(nx.number_of_nodes(G)) for i in degree_values]
plt.figure()
plt.plot(degree_values, histogram, 'o')
plt.title('Gráfica Distribución de grados de Barabasi Albert')
plt.xlabel("Grados")
plt.ylabel("Fracción de nodos")
plt.xscale('log')
plt.yscale('log')
plt.show()
```
